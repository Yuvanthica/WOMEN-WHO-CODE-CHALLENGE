Write a program that uses recursion to generate all permutations of a list

```python
def perm(lst, end=[]):
    if(len(lst) == 0):
        print(end)
    else:
        for i in range(len(lst)):
            perm(lst[:i] + lst[i+1:], end + lst[i:i+1])
```

```python
perm([1,2,3,9,3])
```
